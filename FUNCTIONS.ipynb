```python

def add(x,y):
    return(x+y)
def sub(x,y):
    return(x-y)
def mul(x,y):
    return(x*y)
def div(x,y):
    return(x/y)
   
print("choose 1 for addition")
print("choose 2 for subtraction")
print("choose 3 for multiplication")
print("choose 4 for division")
choose=int(input('choose='))
num1=int(input("enter num1="))
num2=int(input("enter num2=")) 
if(choose==1):
    sum=add(num1,num2)
    print("result =",sum )
elif(choose==2):
    sub=sub(num1,num2)
    print("result =",sub )
elif(choose==3):
    mul=mul(num1,num2)
    print("result =",mul)
elif(choose==4):  
    div=div(num1,num2)
    print("result =",div)
else:
    print("operator not matching")    
```

```python
def add(x,y):
    return(x+y)
def sub(x,y):
    return(x-y)
def mul(x,y):
    return(x*y)
def div(x,y):
    return(x/y)
 
print("choose 1 for addition")
print("choose 2 for subtraction")
print("choose 3 for multiplication")
print("choose 4 for division")
choose=int(input('choice='))
num1=int(input("enter num1="))
num2=int(input("enter num2="))   
if(choose==1):
    print(add(num1,num2))
elif(choose==2):
    print(sub(num1,num2))
elif(choose==3):
    print(mul(num1,num2))
elif(choose==4):  
    print(div(num1,num2))
    
else:
    print("operator not matching")    
```
