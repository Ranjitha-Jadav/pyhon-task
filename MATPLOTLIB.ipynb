```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#SCATTER PLOT
```

```python
#syntx-----plt.scatter()
```

```python
#creating tw0 list

x=np.arange(10)
y=np.arange(10,20)
print(x)
print(y)
```

```python
#to create scatter plot

plt.scatter(x,y,c='g')

#xlabel
plt.xlabel('X values')
#ylabel
plt.ylabel('Y label')
#title of graph
plt.title('scatter plot')

#to see the plot
plt.show()
```

```python
#use of marker
plt.scatter(x,y,c= 'g',marker = '_')
# plt.plot(x,y,'r')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('2d Diagram')
plt.show()
```

```python
#nsyntx---plt.plot()
```

```python
np.pi
```

```python
X1=np.arange(0,2*np.pi,0.1)
Y1=np.sin(X1)
print(X1)
print(Y1)
```

```python
#sine wave from 0 t0 2 pi

plt.plot(X1,Y1,'g')
plt.xlabel('x value')
plt.ylabel('y value')
plt.title('sin wave')
plt.show()
```

```python
# cosine wave from 0 to 2 pie
x1 = np.arange(0,2*np.pi,0.1)
y1 = np.cos(x1)

plt.plot(x1,y1,'r+')
plt.xlabel('X Values')
plt.ylabel('COsine Values')
plt.title('Cosine wave')
plt.show()
```

```python
plt.plot(x1,y1,'bo')
plt.xlabel('X Values')
plt.ylabel('COsine Values')
plt.title('Cosine wave')
plt.show()
```

```python
#PIE CHART
#syntx----plt.pie()
```

```python
#data to plot

x=[100,50,30,10]
labels=['china','us','russia','india']
colors=['r','g','b','y']
explode=(0,0,0.5,0)
#plot
plt.pie(x,labels=labels,colors=colors,explode=explode,shadow=True,autopct='%0.3f%%')
plt.title('corona virus affected people')
plt.show
```

```python
#to creat 4 plots
x=np.arange(10)
y=x**3
print(x)
print(y)
```

```python
#to change the fig size
plt.figure(figsize=(10,10))

#1st plot
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.title('fig 1')

#2nd plot
plt.subplot(2,2,2)
plt.plot(x,y,'g--')
plt.title('fig 2')

#3rd plot
plt.subplot(2,2,3)
plt.plot(x,y,'b*--')
plt.title('fig 3')

#4th plot
plt.subplot(2,2,4)
plt.plot(x,y,'yo')
plt.title('fig 4')
plt.show()
```
